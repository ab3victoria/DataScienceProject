```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt      
import math
import seaborn as sns
from collections import Counter
from sklearn.decomposition import PCA
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data/final.csv')
df.columns
df.drop(columns={'name'},axis=1,inplace=True)
df
```

## Heatmap:

```python
plt.figure(figsize =(10,7))
sns.heatmap(df.corr(), annot = True)
```

```python
plt.title("Correlation between deffensive rebounds and number of field goals",fontdict={'fontname':'Tahoma','fontsize':20})
plt.xlabel('deffensRebounds')
plt.ylabel('fieldGoals')

plt.scatter(df.deffensRebounds, df.fieldGoals)
df['deffensRebounds'].corr(df['fieldGoals'])
```

```python
plt.title("Correlation between offensive rebounds and number of 2 point goals",fontdict={'fontname':'Tahoma','fontsize':20})
plt.xlabel('offensRebounds')
plt.ylabel('point2_Goals')
plt.scatter(df.offensRebounds, df.point3_Goals)
df['offensRebounds'].corr(df['point2_Goals'])
```

```python
df.columns
```

```python
plt.title("Correlation between number of free throws attempts and field goals",fontdict={'fontname':'Tahoma','fontsize':20})
plt.xlabel('freeThrowsAttemp')
plt.ylabel('fieldGoals')
plt.scatter(df.freeThrowsAttemp, df.fieldGoals)
df['freeThrowsAttemp'].corr(df['fieldGoals'])
```

```python
plt.title("Correlation between number of free throws and draft",fontdict={'fontname':'Tahoma','fontsize':20})
plt.xlabel('fieldGoals')
plt.ylabel('draft')
plt.scatter(df.fieldGoals, df.draft)
```

# Linear plot:
#### As we see, the more minutes a player had in a game, the better results he got:

```python
plt.title("Line plot graph - showing minutes played",fontdict={'fontname':'Tahoma','fontsize':20})
ax = sns.lineplot(x='minPlayed',y='fieldGoals',data=df, label='field goals')
ax = sns.lineplot(x='minPlayed',y='point2_Goals',data=df, label='2 goals')
ax = sns.lineplot(x='minPlayed',y='point3_Goals',data=df, label='3 goals')
ax.set(ylabel = 'avg')
ax.set(xlabel = 'minutes')
ax.legend()
plt.show()
```
