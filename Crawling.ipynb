# Crawling Data from "sports-reference" website, using selenium
### We used a function to get our attributes to a dataframe. We collected the attributes by players names in alphabetical order.


```python
PATH = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(PATH)
driver.maximize_window()
main_data = list()

def attribute(driver):
    
    players = driver.find_elements(By.XPATH, '//*[@id="meta"]/div/h1/span')
    g = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[3]')
    gs = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[4]')
    mp = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[5]')
    fg = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[6]')
    fga = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[7]')
    fg1 = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[8]')
    p2 = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[9]')
    p2a = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[10]')
    p21 = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[11]')
    p3 = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[12]')
    p3a = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[13]')
    p31 = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[14]')
    ft = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[15]')
    fta = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[16]')
    ft1 = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[17]')
    orb = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[18]')
    drb = driver.find_elements(By.XPATH, '//*[@id="players_per_game"]/tfoot/tr/td[19]')

    data = []
    i = 0
    

    temporary_data = {'name': players[i].text, 'G': g[i].text, 'GS': gs[i].text,
                      'MP': mp[i].text, 'FG': fg[i].text, 'FGA': fga[i].text,
                      'FG%': fg1[i].text, 'P2': p2[i].text, '2PA': p2a[i].text,
                      '2P%': p21[i].text, '3P': p3[i].text, '3PA': p3a[i].text,
                      '3P%': p31[i].text, 'FT': ft[i].text, 'FTA': fta[i].text,
                      'FT%': ft1[i].text, 'ORB': orb[i].text, 'DRB': drb[i].text,
                      }

    data.append(temporary_data)
    return data


# Empty list for storing links
links = []
final_links =[]
listOfpSmall = []

str1 = ".html"
str2 = "/cbb/players/"

html_text=requests.get('https://www.sports-reference.com/cbb/players/a-index.html').text

soup = BeautifulSoup(html_text,'html.parser')


mtag = soup.find('div',attrs={"id":"content"})

links=[link['href'] for link in mtag.findAll("a")]



for link in links:
    if(link.endswith(str1) & link.startswith(str2)):
        final_links.append(link)
        
for link in final_links:        
        driver.get("https://www.sports-reference.com"+link);
        main_data += attribute(driver)
        
        

df = pd.DataFrame(main_data)        

df.to_csv(r'C:\Users\vikvi\0_NBA_PROJECT\Dataframe_A.csv',index=False)

```

# Crawling from Official NBA website
### We collected information about players picks in first draft - of all times.

```python
PATH = "C:\Program Files (x86)\chromedriver.exe"
driver=webdriver.Chrome(PATH)
driver.get("https://www.nba.com/stats/draft/history/?Season=")
```

```python

players = driver.find_elements(By.XPATH,'/html/body/main/div/div/div[2]/div/div/nba-stat-table/div[2]/div[1]/table/tbody/tr/td[1]')
rpicks = driver.find_elements(By.XPATH,'/html/body/main/div/div/div[2]/div/div/nba-stat-table/div[2]/div[1]/table/tbody/tr/td[5]')
result=[]

for i in range(len(players)):
    temp={'Name':players[i].text,'DraftPick':rpicks[i].text}

    result.append(temp)
    
df=pd.DataFrame(result)


df.to_csv(r'C:\Users\vikvi\0_FINAL\Dataframe_NBA_Correct.csv',index=False)
```
